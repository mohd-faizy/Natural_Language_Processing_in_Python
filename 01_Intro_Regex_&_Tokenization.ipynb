What exactly are Regular Expressions?</span><br>

Regular expressions (often abbreviated as **regex**) are strings with a special syntax that allow us to match patterns in other strings. They are incredibly powerful tools for text processing.

### Applications of Regular Expressions
*   Finding all web links in a document.
*   Parsing email addresses.
*   Removing or replacing unwanted characters.

### Basic Example
Below is a simple example of using the `re` module to match a pattern.



```python
import re

# Matching a specific string literal
# This checks if the string starts with 'abc'
match_obj = re.match('abc', 'abcdef')
print(match_obj)

# Using a special pattern \w+ (matches word characters)
word_regex = '\w+'
match_obj_2 = re.match(word_regex, 'hi there!')
print(match_obj_2)
```

Common Regex Patterns</span><br>

Understanding the syntax is key to mastering regex. Below is a comprehensive table of common patterns used to identify different types of characters.

| Pattern | Matches | Example |
| :--- | :--- | :--- |
| `\w+` | word | 'Magic' |
| `\d` | digit | 9 |
| `\s` | space | ' ' |
| `.*` | wildcard (any character) | 'username74' |
| `+` or `*` | greedy match | 'aaaaaa' |
| `\S` | not space | 'no_spaces' |
| `[a-z]` | lowercase group | 'abcdefg' |



```python
import re

# Example: Finding digits
text = "The year is 2023"
# \d+ means one or more digits
print(re.findall(r'\d+', text))

# Example: Finding non-space characters
text_nospace = "no_spaces"
print(re.findall(r'\S+', text_nospace))
```

Python's re Module</span><br>

Python's built-in `re` module provides several functions to work with regular expressions.

### Key Functions
*   **`split`**: Split a string on a regex pattern.
*   **`findall`**: Find all non-overlapping patterns in a string.
*   **`search`**: Search for a pattern anywhere in the string.
*   **`match`**: Match an entire string or substring based on a pattern (starts at the beginning).

### Syntax Note
The pattern comes first, and the string comes second: `re.function(pattern, string)`. These functions may return an iterator, a string, or a match object.

### Example: `re.split`



```python
import re

# Splitting a string based on whitespace
# \s+ matches one or more whitespace characters
result = re.split('\s+', 'Split on spaces.')
print(result)
```

### What is Tokenization?
Tokenization is the process of turning a string or document into **tokens** (smaller chunks). It is a fundamental step in preparing text for NLP.

### Concepts
*   There are many different theories and rules for tokenization.
*   You can create your own rules using regular expressions.

### Examples of Tokenization
*   Breaking out words or sentences.
*   Separating punctuation.
*   Separating all hashtags in a tweet.

The NLTK Library</span><br>

**NLTK** stands for Natural Language Toolkit. It is one of the most popular libraries for NLP in Python.

To use NLTK's tokenizers, you often need to download specific data packages (like `punkt`).



```python
import nltk
# Ensure the necessary NLTK data is downloaded
try:
    nltk.data.find('tokenizers/punkt')
except LookupError:
    nltk.download('punkt')
    nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize

# Tokenizing a simple sentence
tokens = word_tokenize("Hi there!")
print(tokens)
```

Why Tokenize?</span><br>

Tokenization makes text easier to analyze programmatically.

### Benefits
1.  **Easier to map Part of Speech (POS)**: Identifying nouns, verbs, etc.
2.  **Matching common words**: Grouping words together.
3.  **Removing unwanted tokens**: Filtering out noise.

### Example Scenario
Consider the sentence: `"I don't like Sam's shoes."`

If we split strictly by space, we might get `"don't"` or `"shoes."`. Proper tokenization handles punctuation and contractions intelligently:
*   `"I"`, `"do"`, `"n't"`, `"like"`, `"Sam"`, `"'s"`, `"shoes"`, `"."`

Other NLTK Tokenizers</span><br>

NLTK provides specialized tokenizers for different needs.

*   **`sent_tokenize`**: Tokenize a document into sentences.
*   **`regexp_tokenize`**: Tokenize a string or document based on a user-defined regular expression pattern.
*   **`TweetTokenizer`**: A special class designed for tweet tokenization. It allows you to separate hashtags, mentions, and handle multiple exclamation points (e.g., `!!!`).



```python
from nltk.tokenize import sent_tokenize, TweetTokenizer

# Sentence Tokenization
text = "Hello world. This is a second sentence."
print("Sentences:", sent_tokenize(text))

# Tweet Tokenization
tweet = "This is a cooool #dweet! :)"
tknzr = TweetTokenizer()
print("Tweet Tokens:", tknzr.tokenize(tweet))
```

More Regex Practice: Search vs. Match</span><br>

It is crucial to understand the difference between `re.search()` and `re.match()`.

*   **`re.match()`**: Tries to match the pattern **only at the beginning** of the string.
*   **`re.search()`**: Scans through the **entire string** looking for the first location where the pattern produces a match.



```python
import re

# Scenario 1: Pattern is at the start
# Both match and search will find it.
print("--- 'abc' in 'abcde' ---")
print("Match:", re.match('abc', 'abcde'))
print("Search:", re.search('abc', 'abcde'))

# Scenario 2: Pattern is NOT at the start
# match will fail (return None), search will succeed.
print("\n--- 'cd' in 'abcde' ---")
print("Match:", re.match('cd', 'abcde'))
print("Search:", re.search('cd', 'abcde'))
```

Advanced Tokenization with Regex</span><br>

You can build complex tokenizers using regex groups and logical operators.

### Key Concepts
*   **OR Operator (`|`)**: Matches either the pattern on the left or the right.
*   **Groups (`()`)**: Defines a group to treat multiple characters as a single unit.
*   **Explicit Character Ranges (`[]`)**: Defines a set of allowed characters.

### Example: Digits OR Words
The pattern `(\d+|\w+)` looks for a sequence of digits OR a sequence of word characters.



```python
import re

match_digits_and_words = '(\d+|\w+)'
text = 'He has 11 cats.'

# findall returns all non-overlapping matches
tokens = re.findall(match_digits_and_words, text)
print(tokens)
```

Here is a table of advanced patterns using ranges and groups.

| Pattern | Matches | Example |
| :--- | :--- | :--- |
| `[A-Za-z]+` | upper and lowercase English alphabet | 'ABCDEFghijk' |
| `[0-9]` | numbers from 0 to 9 | 9 |
| `[A-Za-z\-\.]+` | upper/lowercase English alphabet, `-` and `.` | 'My-Website.com' |
| `(a-z)` | a, `-` and z | 'a-z' |
| `(\s+\|,)` | spaces or a comma | ', ' |

<div style="background: #e0f2fe; border-left: 16px solid #0284c7; padding: 14px 18px; border-radius: 8px; font-size: 18px; color: #075985;"> ðŸ’¡ <b>Tip:</b> Inside square brackets <code>[]</code>, the hyphen <code>-</code> is used to define a range (e.g., a-z). If you want to match a literal hyphen, you often need to escape it <code>\-</code> or place it at the end of the group. </div>

Let's practice using a character range to match a complex string.



```python
import re

my_str = 'match lowercase spaces nums like 12, but no commas'

# Pattern explanation:
# [a-z0-9 ]+ matches one or more lowercase letters, numbers, or spaces.
# It will stop matching when it hits the comma.
match_obj = re.match('[a-z0-9 ]+', my_str)

print(match_obj)
# Notice the span ends at 42, before the comma
```

Combining NLP Data Extraction with Plotting</span><br>

Now, let's put it all together. We will:
1.  Tokenize a string using NLTK.
2.  Calculate the length of each word.
3.  Plot the distribution of word lengths using Matplotlib.



```python
from matplotlib import pyplot as plt
from nltk.tokenize import word_tokenize

# 1. Tokenize
text = "This is a pretty cool tool!"
words = word_tokenize(text)

# 2. Extract lengths using a list comprehension
word_lengths = [len(w) for w in words]
print("Words:", words)
print("Lengths:", word_lengths)

# 3. Plot
plt.hist(word_lengths)
plt.title("Word Length Histogram")
plt.xlabel("Length of Word")
plt.ylabel("Frequency")
plt.show()
```
